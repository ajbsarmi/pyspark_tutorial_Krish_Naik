# Adding Columns

```python
# Add columns in a dataframe
df_pyspark = df_pyspark.withColumn(
    "Experience",
    F.when(F.col("Name") == "Arden Sarmiento", 4)
     .when(F.col("Name") == "Mabel Sarmiento", 6)
     .when(F.col("Name") == "Aerold Sarmiento", 8)
)
df_pyspark.show()
```

```python
df_pyspark = df_pyspark.withColumn('Experience After 2 years',df_pyspark['Experience']+2)
```

```python
df_pyspark
```

```python
df_pyspark.show()
```

# Drop the columns

```python
# Drop the columns
df_pyspark = df_pyspark.drop('Experience After 2 years')
```
